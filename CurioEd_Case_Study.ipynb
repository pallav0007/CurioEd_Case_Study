```python
import matplotlib.pyplot as plt
import numpy as np
```

###Information and Assumption initialization

```python
#information

#mentor
mentor_absent=5
mentor_late=10

#student
student_absent=30
student_late=20

#assumptions
students=1000
Faculty=85
slot_per_faculty=4
working_days=24
gap_days=3
duration=1
avg_time_wasted=8/60
```

```python
def student_behaviour(absent,late):
  remaining=100-absent-late
  y = np.array([absent,late,remaining])
  mylabels = [f"Absent {absent}%", f"Late {late} %", f"Remaining {remaining}%"]
  myexplode = [0.2, 0.1, 0.1]

  plt.pie(y, labels = mylabels,explode=myexplode)
  plt.legend(title = "Student Behaviour @Curioed")
  plt.savefig("sb.png")
student_behaviour(student_absent,student_late)
```

```python
def mentor_behaviour(absent,late):
  remaining=100-absent-late
  y = np.array([absent,late,remaining])
  mylabels = [f"Absent {absent}%", f"Late {late} %", f"Remaining {remaining}%"]
  mycolors = ["red", "yellow", "#4CAF50"]
  myexplode = [0,0,0]
  plt.pie(y, labels = mylabels,explode=myexplode,colors=mycolors)
  plt.legend(title = "Mentor Behaviour @CurioEd")
  plt.savefig("mb.png")
mentor_behaviour(mentor_absent,mentor_late)
```

```python
def time_wastage():
  class_per_day=Faculty*slot_per_faculty
  student_per_day=class_per_day
  absent_student=student_per_day*30*1/100
  absent_mentor=class_per_day*mentor_absent/100
  student_time_waste=avg_time_wasted*student_per_day*student_late/100
  mentor_time_waste=avg_time_wasted*class_per_day*mentor_late/100
  x=["student absent","student late", "mentor absent", "mentor late"]
  y=[absent_student,student_time_waste,absent_mentor,mentor_time_waste]
  plt.bar(x,y, color ='maroon',)
 
  plt.xlabel("caused by")
  plt.ylabel("time in hours")
  plt.title("Time wasted in hours per day @CurioEd")
  plt.savefig("tw.png")
time_wastage()
```

```python
def time_wastage_comparison():
  class_per_day=Faculty*slot_per_faculty
  student_per_day=class_per_day

  #before solving problem statement
  absent_student=student_per_day*30*1/100
  absent_mentor=class_per_day*mentor_absent/100
  student_time_waste=avg_time_wasted*student_per_day*student_late/100
  mentor_time_waste=avg_time_wasted*class_per_day*mentor_late/100


  #after solving problem statement
  after_absent_student=student_per_day*10*1/100
  after_absent_mentor=class_per_day*mentor_absent/100*0
  after_student_time_waste=avg_time_wasted*student_per_day*student_late/100*0
  after_mentor_time_waste=avg_time_wasted*class_per_day*mentor_late/100


  x=["student absent","student late", "mentor absent", "mentor late"]
  y=[absent_student,student_time_waste,absent_mentor,mentor_time_waste]

  y2=[after_absent_student,after_student_time_waste,after_absent_mentor,after_mentor_time_waste]
  print(y2)
  plt.scatter(x,y,color ='red',label=f"{round(sum(y),2)} hrs before optimization")
  plt.scatter(x,y2,color="green",label=f"{round(sum(y2),2)} hrs after optimization")
  plt.legend(title = "Total wastage")
  plt.xlabel("caused by")
  plt.ylabel("time in hours")
  plt.title("Time wasted in hours per day @CurioEd")
  plt.savefig("conclusion.png")
time_wastage_comparison()
```
